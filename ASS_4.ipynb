```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def add_gaussian_noise(img, mean=0, var=30):
    row, col = img.shape
    sigma = var ** 0.5
    gauss = np.random.normal(mean, sigma, (row, col))
    noisy = img + gauss
    noisy = np.clip(noisy, 0, 255)
    return noisy.astype(np.uint8)

# Read grayscale image
img = cv2.imread('/content/fdip pic.jpg', cv2.IMREAD_GRAYSCALE)
gaussian_img = add_gaussian_noise(img)

# Create smaller figure
plt.figure(figsize=(6, 3))  # reduce overall size (width=6, height=3)

# Plot Gaussian noisy image
plt.subplot(1, 2, 1)
plt.imshow(gaussian_img, cmap='gray')
plt.title("Gaussian Noisy Image", fontsize=8)
plt.axis("off")

# Plot histogram
plt.subplot(1, 2, 2)
plt.hist(gaussian_img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

def add_salt_pepper_noise(img, salt_prob=0.01, pepper_prob=0.01):
    noisy = img.copy()
    row, col = img.shape

    # Salt noise
    num_salt = int(salt_prob * row * col)
    for i in range(num_salt):
        y = random.randint(0, row - 1)
        x = random.randint(0, col - 1)
        noisy[y, x] = 255

    # Pepper noise
    num_pepper = int(pepper_prob * row * col)
    for i in range(num_pepper):
        y = random.randint(0, row - 1)
        x = random.randint(0, col - 1)
        noisy[y, x] = 0

    return noisy

# Read grayscale image
img = cv2.imread('/content/fdip pic.jpg', cv2.IMREAD_GRAYSCALE)
sp_img = add_salt_pepper_noise(img)

# Create smaller figure
plt.figure(figsize=(6, 3))  # shrink display size (width=6, height=3)

# Plot Salt & Pepper noisy image
plt.subplot(1, 2, 1)
plt.imshow(sp_img, cmap='gray')
plt.title("Salt & Pepper Noisy Image", fontsize=8)
plt.axis("off")

# Plot histogram
plt.subplot(1, 2, 2)
plt.hist(sp_img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

def add_impulse_noise(img, prob=0.02):
    noisy = img.copy()
    row, col = img.shape
    num_noise = int(prob * row * col)

    for i in range(num_noise):
        y = random.randint(0, row - 1)
        x = random.randint(0, col - 1)
        noisy[y, x] = random.choice([0, 255])

    return noisy

# Read grayscale image
img = cv2.imread('/content/fdip pic.jpg', cv2.IMREAD_GRAYSCALE)
impulse_img = add_impulse_noise(img)

# Create smaller figure
plt.figure(figsize=(6, 3))  # reduce output size (width=6, height=3)

# Plot impulse noisy image
plt.subplot(1, 2, 1)
plt.imshow(impulse_img, cmap='gray')
plt.title("Impulse Noisy Image", fontsize=8)
plt.axis("off")

# Plot histogram
plt.subplot(1, 2, 2)
plt.hist(impulse_img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram", fontsize=8)

plt.tight_layout()
plt.show()
```
